Convert the ordinal data into dummy data

The idea is that

Step 1, taking out all the categorical data and convert them into dummy variables.<br>
Step 2, concatenating the numerical and dummy variables together.

```python
# converting all the categorical variables to dummy variables
bankCategorical = pd.get_dummies(bankData[['job', 'marital', 'education', 'default', 'housing', 'loan','contact', 'month', 'poutcome']])
print(bankCategorical.shape)
bankCategorical.head()
```

```python
# extracting the numerical data
bankNum = bankData[['age', 'balance', 'day','duration','campaign','pdays', 'previous']]
bankNum.shape
```

```python
# concatenating them together and prepare the X values

X = pd.concat([bankCategorical, bankNum], axis=1)
print(X.shape)
# prepare y value
Y = bankData['y']
print(Y.shape)
```
